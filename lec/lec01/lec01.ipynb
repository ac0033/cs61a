```python
# Objects
# Note: Download from http://composingprograms.com/shakespeare.txt
shakes = open('shakespeare.txt')
text = shakes.read().split()
print(len(text))
text[:25]
```

```python
print(text.count('our'), 
text.count('thou'), 
text.count('you'), 
text.count('forsooth'), 
text.count('the'), 
text.count('.'), 
text.count(':'), 
text.count(','), 
text.count(',') / len(text))
```

```python
# Sets
words = set(text)
len(words)
```

```python

# Combinations and Data 
{w for w in words if len(w) == 4}
```
